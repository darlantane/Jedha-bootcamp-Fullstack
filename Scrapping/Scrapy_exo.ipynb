```python
import pandas as pd
import os
import logging
import scrapy
from scrapy.crawler import CrawlerProcess
```

```python
class spectacles_(scrapy.Spider):
    name = "spectacles_"
    start_urls = ["http://offi.fr/theatre/programme.html"]

    def parse(self, response):
        spectacles = response.css("div.column")

        for spectacle in spectacles:
            yield {
                'titre': spectacle.css('div.event-title h3.mb-0 a.text-body span::text').get()
            }
```

```python
filename = "spectacles.json"

if filename in os.listdir("src/"):
    os.remove("src/" + filename)
```

```python
process = CrawlerProcess(settings={
    'USER_AGENT': 'Mozilla/4.0',
    'LOG_LEVEL':logging.INFO,
    "FEEDS":{
        "src/" + filename : {"format":"json"}
    }
}
    
)

process.crawl(spectacles_)
process.start()
```
