```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df_arbres = pd.read_excel('les-arbres.xlsx')
df_arbres.describe(include = 'all')
```

```python
x = [1, 2, 3, 4, 5, 6]

y = [12, 9, 20, 8, 6, 2]
z = [9, 12, 18, 3, 4, 2]

plt.plot(x, y)
#plt.plot(x, z, linestyle = "--")
plt.plot(x, z, "o")


# titre
plt.title(label = "Mon premier graphique",
          fontsize = 18)

# titre des axes
plt.xlabel("Mon axe x")
plt.ylabel("Mon axe y")

# légende
plt.legend(["Mes données Y", "Mes données Z"])

plt.show()
```

```python
df_especes = df_arbres.groupby('LIBELLE FRANCAIS').count().reset_index()
df_especes
```

```python
N=10
df_especes_top = df_especes.sort_values(by='IDBASE', ascending = False).head(N)

x= df_especes_top['LIBELLE FRANCAIS']
y= df_especes_top['IDBASE']

plt.bar(x, y)
plt.xticks(rotation='vertical')
plt.title(f'Top {N} espèces d_arbres à Paris')

for i, valeur in enumerate(df_especes_top['IDBASE']):
    plt.text(i, valeur + 500 , valeur, ha='center', fontsize=8)

plt.show()
```

```python
df_arro = df_especes = df_arbres.groupby('ARRONDISSEMENT').count().reset_index()
df_arro
```

```python
N = 10
df_arro_top = df_arro.sort_values(by='IDBASE', ascending=False).head(N)

x = df_arro_top['ARRONDISSEMENT']
y = df_arro_top['IDBASE']

plt.title(f'Le top {N} des arrondissements en matière d_arbres')


plt.barh(x, y)

plt.show()
```

```python
fig, axes = plt.subplots(2,3)
fig.set_size_inches(15,12)
fig.tight_layout(pad=3.0)


N = 10

### Créer un dataframe qui contient le nb d'arbres par espèces
df_especes = df_arbres.groupby("LIBELLE FRANCAIS")['IDBASE'].count().reset_index()

# Création d'un dataframe interemédiaire avec le top 10 des espèces en ordre décroissant
df_especes_top = df_especes.sort_values(by = 'IDBASE',
                       ascending = False ).head(N)

x = df_especes_top['LIBELLE FRANCAIS']
y = df_especes_top['IDBASE']

axes[0,0].bar(x, y)






df_stade = df_arbres.groupby("STADE DE DEVELOPPEMENT")['IDBASE'].count().reset_index()


x = df_stade["STADE DE DEVELOPPEMENT"]
y = df_stade["IDBASE"]

axes[0,1].pie(y,
        labels = x,
        explode = [0, 0, 0, 0.2],
        autopct="%1.1f%%" )


# Rotation des labels de l'axe x
axes[0, 0].tick_params(axis='x', labelrotation=90, size = 10)

#Titre du graphique et des axes
axes[0,0].set_title(f"Top {N} espèces d'arbres à Paris")
axes[0,0].set_ylabel("Nombre")

plt.show()
```
