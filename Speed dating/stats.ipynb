```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Speed+Dating+Data.csv', encoding='cp1252')
data.head()
```

```python
data.columns
```

```python
ds=data[['gender','age','race','imprace','imprelig']]
ds.head()
```

```python
race_labels = {1.0 : 'Black/African American',
	2.0 : 'European/Caucasian-American',
	3.0 : 'Latino/Hispanic American',
	4.0 : 'Asian/Pacific Islander/Asian-American',
	5.0 : 'Native American=5',
	6.0 : 'Other'
}

gender_labels = {0 : 'Female', 1 : 'Male'}

ds['race'] = ds['race'].replace(race_labels)
ds['gender']= ds['gender'].replace(gender_labels)
```

```python
plt.figure(figsize=(7, 7))
plt.xticks(rotation='vertical')
plt.title("Importance de la race selon la race et le genre")

sns.barplot(ds, x='race', y='imprace', hue='gender', errorbar=None, palette='rocket')
```

```python
plt.figure(figsize=(15, 6))
plt.title('Importance de la race selon l_age')
sns.barplot(ds, x='age', y='imprace', hue='gender', errorbar=None, palette='pastel')
```

```python
plt.figure(figsize=(7, 7))
plt.xticks(rotation='vertical')
plt.title("Importance de la religion selon la race et le genre")

sns.barplot(ds, x='race', y='imprelig', hue='gender', errorbar=None, palette='rocket')
```

```python
plt.figure(figsize=(15, 6))
plt.title('Importance de la religion selon l_age')
sns.barplot(ds, x='age', y='imprelig', hue='gender', errorbar=None, palette='pastel')
```

```python
dt = data[['gender','age','race','match','sports','tvsports','exercise','dining','museums','art','hiking','gaming','clubbing','reading','tv','theater','movies','concerts','music','shopping','yoga']]
dt
```

```python
lower_bound = dt.quantile(0.01)
upper_bound = dt.quantile(0.99)
dt = dt[(dt >= lower_bound) & (dt <= upper_bound)]

dt['race'] = dt['race'].replace(race_labels)
dt['gender']= dt['gender'].replace(gender_labels)
```

```python
plt.figure(figsize=(5, 5))
plt.xticks(rotation='vertical')
plt.title("Taux de match des fans de sport")

sns.barplot(dt, x='tvsports', y='match', hue='gender', errorbar=None, palette='rocket')
```
