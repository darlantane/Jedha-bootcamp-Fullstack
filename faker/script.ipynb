```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
produits = pd.read_csv('src/Produits.csv')
clients = pd.read_csv('src/Clients.csv')

produits = produits.rename(columns={'sum(paiement.montant)': 'valeur'})
produits
```

```python
plt.figure(figsize=(10, 4))
plt.title('Les produits les plus rentables')
plt.xticks(rotation='vertical')
sns.barplot(produits, x='nom', y='valeur', errorbar=None, palette = 'rocket')
```

```python
clients = clients.rename(columns={'sum(paiement.montant)': 'valeur'})
clients2 = pd.DataFrame(clients.groupby('ville')['valeur'].mean())
clients2
```

```python
colors = sns.color_palette("pastel")
labels = clients['ville'].unique()
plt.pie(clients.groupby('ville')['valeur'].mean(), labels=labels, autopct="%0.0f%%", colors=colors,
        textprops={"color": "Green"})
```
